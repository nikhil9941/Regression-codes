```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Importing the dataset
data_set = pd.read_csv('path/to/your/file/50_CompList.csv')  # Update the path if needed

# Extracting Independent (X) and Dependent (y) Variables
x = data_set.iloc[:, :-1].values
y = data_set.iloc[:, -1].values  # Assuming the dependent variable is in the last column

# Checking the shape of X to ensure the index is valid
print("Shape of X:", x.shape)

# Identify the column index that needs encoding
categorical_column_index = 3  # Adjust this index based on your dataset

if categorical_column_index < x.shape[1]:
    # Encoding the categorical data in the specified column
    labelencoder_x = LabelEncoder()
    x[:, categorical_column_index] = labelencoder_x.fit_transform(x[:, categorical_column_index])

    # Using ColumnTransformer to create dummy variables for categorical data
    ct = ColumnTransformer([("State", OneHotEncoder(), [categorical_column_index])], remainder='passthrough')
    x = ct.fit_transform(x)

    # Avoiding the dummy variable trap by dropping the first column of dummy variables
    x = x[:, 1:]
else:
    print(f"Index {categorical_column_index} is out of bounds for axis 1 with size {x.shape[1]}.")

# Splitting the dataset into the training set and the test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Fitting the Multiple Linear Regression model to the training set
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(x_test)

# Printing the training and testing scores
print('Train Score:', regressor.score(x_train, y_train))
print('Test Score:', regressor.score(x_test, y_test))

# Plotting the Predicted vs. Actual Values
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)  # Line for perfect prediction
plt.title('Actual vs Predicted')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()

# Plotting Residuals (Error) vs. Predicted Values
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, color='blue')
plt.axhline(y=0, color='red', linewidth=2)
plt.title('Residuals vs Predicted Values')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

# Plotting the Distribution of Residuals
sns.histplot(residuals, kde=True)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.show()
```
